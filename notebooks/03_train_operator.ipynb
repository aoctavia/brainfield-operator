```python
from brainfield_operator.data import BrainFieldDataset
from brainfield_operator.models import FNO2d
from brainfield_operator.training import fit_operator_model

data_dir = "../data/brainfield"

dataset = BrainFieldDataset(root_dir=data_dir)
n_total = len(dataset)
n_val = int(0.2 * n_total)
n_train = n_total - n_val

from torch.utils.data import random_split
train_dataset, val_dataset = random_split(dataset, [n_train, n_val])

model = FNO2d(in_channels=2, out_channels=1, width=64, modes_x=16, modes_y=16)

fit_operator_model(
    model=model,
    train_dataset=train_dataset,
    val_dataset=val_dataset,
    batch_size=8,
    lr=1e-3,
    num_epochs=50,
    device="cuda",
    save_dir="../checkpoints",
    seed=42,
)
```
